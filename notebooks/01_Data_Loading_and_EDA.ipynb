This lets us process every dataset with a loop.

```python
datasets = {
    "tasks": tasks,
    "employees": employees,
    "skills": skills,
    "workload": workload,
    "comments": comments,
    "attachments": attachments,
    "task_history": task_history,
    "clients": clients,
    "projects": projects,
    "employee_performance": employee_performance,
    "sprints": sprints
}
```

Check Duplicate Records

```python
for name, df in datasets.items():
    print(f"{name}: {df.duplicated().sum()} duplicate rows")
```

```python
for name, df in datasets.items():
    original = globals()[name].shape[0]
    cleaned = datasets[name].shape[0]

    print(f"{name}")
    print(f"Before : {original}")
    print(f"After  : {cleaned}")
    print("-"*30)
```

```python
datasets["tasks"].to_csv("../data/processed/tasks_clean.csv", index=False)
datasets["employees"].to_csv("../data/processed/employees_clean.csv", index=False)
datasets["skills"].to_csv("../data/processed/skills_clean.csv", index=False)
datasets["workload"].to_csv("../data/processed/workload_clean.csv", index=False)
datasets["comments"].to_csv("../data/processed/comments_clean.csv", index=False)
datasets["attachments"].to_csv("../data/processed/attachments_clean.csv", index=False)
datasets["task_history"].to_csv("../data/processed/task_history_clean.csv", index=False)
datasets["clients"].to_csv("../data/processed/clients_clean.csv", index=False)
datasets["projects"].to_csv("../data/processed/projects_clean.csv", index=False)
datasets["employee_performance"].to_csv("../data/processed/employee_performance_clean.csv", index=False)
datasets["sprints"].to_csv("../data/processed/sprints_clean.csv", index=False)
```
