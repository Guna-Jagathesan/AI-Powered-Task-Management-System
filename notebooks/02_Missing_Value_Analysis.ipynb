```python
import pandas as pd
import numpy as np

tasks = pd.read_csv("../data/processed/tasks_clean.csv")
employees = pd.read_csv("../data/processed/employees_clean.csv")
skills = pd.read_csv("../data/processed/skills_clean.csv")
workload = pd.read_csv("../data/processed/workload_clean.csv")
comments = pd.read_csv("../data/processed/comments_clean.csv")
attachments = pd.read_csv("../data/processed/attachments_clean.csv")
task_history = pd.read_csv("../data/processed/task_history_clean.csv")
clients = pd.read_csv("../data/processed/clients_clean.csv")
projects = pd.read_csv("../data/processed/projects_clean.csv")
employee_performance = pd.read_csv("../data/processed/employee_performance_clean.csv")
sprints = pd.read_csv("../data/processed/sprints_clean.csv")
```

Store All DataFrames

```python
datasets = {
    "tasks": tasks,
    "employees": employees,
    "skills": skills,
    "workload": workload,
    "comments": comments,
    "attachments": attachments,
    "task_history": task_history,
    "clients": clients,
    "projects": projects,
    "employee_performance": employee_performance,
    "sprints": sprints
}
```

Check Missing Values

```python
for name, df in datasets.items():
    print("="*30)
    print(f"Dataset: {name}")
    print(df.isnull().sum())
```

Count Total Missing Values

```python
for name, df in datasets.items():
    total_missing = df.isnull().sum().sum()

    print(f"{name}: {total_missing} missing values")
```

Missing Value Percentage

```python
for name, df in datasets.items():

    missing = df.isnull().sum()

    percentage = (missing / len(df))*100

    result = pd.DataFrame({
        "Missing Values": missing,
        "Percentage": percentage
    })

    result = result[result["Missing Values"]>0]

    if len(result)>0:
        print("\n",name)
        display(result)
    else:
        print("No missing values in", name)
```

```python
for name, df in datasets.items():
    print(name, df.isnull().sum().sum())
```

```python
for name, df in datasets.items():
    df.to_csv(f"../data/processed/{name}_processed.csv", index=False)
```
